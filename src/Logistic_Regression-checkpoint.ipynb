```python
data: pd.DataFrame = pd.read_csv("C:/Users/Pikmi/OneDrive/Desktop/CS5402/creditcard.csv/creditcard.csv")
data["Class"] = np.where(data["Class"] <= 0, 0, 1)

data.shape
```

```python

# Split data into training/testing
X = data.drop("Class", axis=1).to_numpy()
Y = data["Class"].to_numpy()

# Split the data into training and testing sets (80% train, 20% test)
np.random.seed(42)  # For reproducibility
training_data_mask = np.random.rand(len(data)) < 0.8
X_train = X[training_data_mask]
Y_train = Y[training_data_mask]
X_test = X[~training_data_mask]
Y_test = Y[~training_data_mask]


```

Logistic Regression

```python

# Implement Logistic Regression
def sigmoid(z):
    # Use np.clip to prevent overflow
    z = np.clip(z, -500, 500)  # Clip values to avoid overflow
    return 1 / (1 + np.exp(-z))

def fit(X, Y, learning_rate=0.01, epochs=500):
    num_samples, num_features = X.shape
    weights = np.zeros(num_features)
    bias = 0.0

    for epoch in range(epochs):
        linear_model = np.dot(X, weights) + bias
        y_predicted = sigmoid(linear_model)

        # Compute gradients
        dw = (1 / num_samples) * np.dot(X.T, (y_predicted - Y))
        db = (1 / num_samples) * np.sum(y_predicted - Y)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

        # Clip predictions to avoid log(0)
        y_predicted = np.clip(y_predicted, 1e-15, 1 - 1e-15)

        if epoch % 50 == 0 or epoch == epochs - 1:
            loss = - (1 / num_samples) * np.sum(Y * np.log(y_predicted) + (1 - Y) * np.log(1 - y_predicted))
            print(f"Epoch {epoch}, Loss = {loss}")

    return weights, bias

def predict(X, weights, bias):
    linear_model = np.dot(X, weights) + bias
    y_predicted = sigmoid(linear_model)
    y_predicted_class = [1 if i > 0.5 else 0 for i in y_predicted]
    return np.array(y_predicted_class)

# Train the model
w, b = fit(X_train, Y_train, learning_rate=0.001, epochs=500)

# Test the model
Y_pred = predict(X_test, w, b)

# Calculate accuracy
accuracy = np.mean(Y_pred == Y_test)
print(f"Accuracy: {accuracy}")

# Output the weights and bias
print(f"Weights: {w}")
print(f"Bias: {b}")
```
